```python
import pandas as pd

# Read the datasets into Pandas DataFrames
dftrain = pd.read_csv('D:\XGBM & LGBM\\Titanic_train.csv')
dftrain 
```

```python
dftrain["New column"]=1  
dftrain
```

```python
dftest=pd.read_csv("D:\XGBM & LGBM\\Titanic_test.csv")
dftest
```

```python
dftest["New column"]=2
dftest
```

```python
dftrain.isnull().sum() ##Finding null values
```

```python
#Filling the null values
dftrain.fillna(value=dftrain['Age'].median(),inplace=True)
dftrain.fillna(value=dftrain['Cabin'].mode(),inplace=True)
dftrain.fillna(value=dftrain['Embarked'].mode(),inplace=True)
dftrain.isnull().sum()
```

```python
dftest.isnull().sum()
```

```python
#Filling the values
dftest.fillna(value=dftest['Age'].median(),inplace=True)
dftest.fillna(value=dftest['Fare'].median(),inplace=True)
dftest.fillna(value=dftest['Cabin'].mode(),inplace=True)
dftest.isnull().sum()
```

```python
df=pd.concat([dftrain,dftest]) ##Concatinating of two dataframes
df
```

```python
df.isnull().sum() 
```

```python
df.hist() ##Histogram
```

```python
import matplotlib.pyplot as plt  ## Boxplot
num_col=df[["Pclass","Age","SibSp","Parch","Fare"]]
plt.boxplot(num_col)
```

```python
###Finding the outliers 
Q1 = num_col.quantile(0.25)   # Calculate Q1 (25th percentile) and Q3 (75th percentile)
Q3 = num_col.quantile(0.75)
IQR = Q3 - Q1  # Calculate the IQR
outliers = ((num_col < (Q1 - 1.5 * IQR)) | (num_col > (Q3 + 1.5 * IQR))).any(axis=1) # Define the outlier criteria
outliers_df=df[outliers]   # identyfying outliers 
outliers_df
```

```python
# Replacing outliers with the median of the respective column
for column in num_col.select_dtypes(include=["float64", "int64"]):
    median_value = num_col[column].median()
    num_col.loc[outliers, column] = median_value

df
```

```python
from sklearn.preprocessing import LabelEncoder  ##Label Encoding
LE=LabelEncoder()
df["Sex"]=LE.fit_transform(df["Sex"])
df['Embarked'] = df['Embarked'].astype(str)
df["Embarked"]=LE.fit_transform(df["Embarked"])
df['Ticket']=LE.fit_transform(df['Ticket'])
df['Cabin'] = df['Cabin'].astype(str)
df['Cabin']=LE.fit_transform(df['Cabin'])
df
```

```python
#Standardizing the continuous data
from sklearn.preprocessing import StandardScaler
ss=StandardScaler()
df[['Pclass']]=ss.fit_transform(df[['Pclass']])
df[['Age']]=ss.fit_transform(df[['Age']])
df[['SibSp']]=ss.fit_transform(df[['SibSp']])
df[['Parch']]=ss.fit_transform(df[['Parch']])
df[['Fare']]=ss.fit_transform(df[['Fare']])
df
```

```python
df_train= df[df['New column'] == 1]
df_train
```

```python
df_test= df[df['New column'] == 2]
df_test=df_test.drop(columns=["Survived"])
df_test
```

```python
#Assigning the X and Y values
Y=dftrain['Survived'].astype(int)
X_train=df_train[['Pclass','Sex','Age','SibSp','Parch','Ticket','Fare','Cabin','Embarked']]
X_test=df_test[['Pclass','Sex','Age','SibSp','Parch','Ticket','Fare','Cabin','Embarked']]
```
